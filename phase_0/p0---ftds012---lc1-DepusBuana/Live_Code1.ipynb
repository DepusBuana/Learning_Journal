#import libraries

````python
import pandas as pd
import numpy as np
````

#Import data

````python
df = pd.read_excel('NMC1-9.xlsx')
df.head()
````

Data berhasil di import dengan keterangan kolom:
time(s): Detik ke- pengukuran
V_m: Voltage/tegangan yang terukur tiap waktu
Current (A): Kuat arus terukur tiap waktu
Cycle: Nomor siklus charge-discharge baterai
Cap: Kapasitas baterai tiap waktu -----> yang akan menjadi parameter acuan perhitungan R/P

#a. make 2 function for R and P

````python
def Hambatan_R(Volt_V,Ampere_A):
    return(Volt_V/Ampere_A)

def Daya_P(Volt_V,Ampere_A):
    return(Volt_V*Ampere_A)
````

#b. Buat kolom baru dengan nama kolom R/P berisikan hasil perhitungan daya/hambatan dengan kriteria: 

Kapasitas >=2.5 hitung daya

Kapasitas <2.5 hitung nilai hambatan. 

Membuat kolom baru untuk hasil tambahan + tambahankan keterangan parameter yang dihitung (untuk kepastian saya saja)

````python
Kolom_hasil = []
Kolom_parameter =[]

for idx in range(0,len(df)):
    
    if df.iloc[idx,4] >= 2.5: #Kapasitas >=2.5 hitung daya
        hasil = Daya_P(df.iloc[idx,1],df.iloc[idx,2])
        Kolom_parameter.append('Daya')

    elif df.iloc[idx,4] < 2.5: #Kapasitas <2.5 hitung nilai hambatan
        hasil = Hambatan_R(df.iloc[idx,1],df.iloc[idx,2])
        Kolom_parameter.append('Hambatan')


    else:
        hasil = 'NaN' #Apabila ada error maka hasil adalah 'NaN'
    
    Kolom_hasil.append(hasil)

df['R/P'] = Kolom_hasil #menambahkan kolom R/P
df['Parameter yang dihitung'] = Kolom_parameter #menambahkan kolom Parameter yg dihitung sebagai acuan
df.head()
````

Note: Cap: Kapasitas baterai tiap waktu -----> yang akan menjadi parameter acuan perhitungan R/P
semua nilai Kapasitas (Cap) diatas 2.5 ( >= 2.5 ) sehingga semua parameter yang dihitung merupakan Daya
dibuktikan oleh:

````python
df.sort_values(by='Cap', ascending=True).head() #sortir secara ascending untuk melihat nilai Cap terkecil
````

Nilai terkecil kolom 'Cap' yaitu 2.77018 dimana 2.77018 > 2.5 sehingga semua parameter yang dihitung merupakan Daya

#c. Hitunglah rata-rata nilai kapasitas untuk arus = 1 Ampere dan cycle ganjil.

Rename Current (A) to Current_A agar bisa menggunakan query

````python
df.rename(columns={'Current (A)':'Current_A'}, inplace = True)
df.head()
````

Column renamed, continue in making querry with constrain:
Current_A == 1 and the cycle is odd (cycle %2 != 0)

````python
df.query("Current_A == 1 & \
Cycle % 2 != 0 "
)
````

Recheck data if all odd cycle already included in the query

````python
df.query("Current_A == 1 & \
Cycle % 2 != 0 "
).groupby(by = ['Current_A','Cycle'])[['Cap']].mean()
````

All data from cycle 1,3,5,7,9 (all odd) has been included. Proceed to calculate mean of capacity (Cap).

````python
df.query("Current_A == 1 & \
Cycle % 2 != 0 "
)[['Cap']].mean()
````

The mean of capacity with constraints as follows: Current_A == 1 and the cycle is odd (cycle %2 != 0)
is 2.806485
